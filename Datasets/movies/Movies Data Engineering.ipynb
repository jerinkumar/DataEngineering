```python
#MISSING DATA ASSESSMENT
```

```python
missing_summary = df.isnull().sum()
print(missing_summary)
```

```python
# check for empty strings and unkown values
for col in df.select_dtypes(include=['object']).columns:
    empty_count = (df[col] == '' ).sum()
    unknown_count = df[col].str.contains('unknown', case=False, na=False).sum() if df[col].dtype == 'object' else 0
    if empty_count > 0 or unknown_count > 0:
        print(f"{col}: {empty_count} empty strings, {unknown_count} unknown values")
```

```python
#Handle missing data
df_copy = df.copy()
```

```python
#Replace missing data with NAN
df_clean = df_copy.replace('', np.nan)
```

```python
#Replace unkown with NaN
df_clean = df_clean.replace('unknown', np.nan, regex=True)
```

```python
print("\n After converting empty strings and unkown to Nan")
print(df_clean['MOVIES'].str.contains('unknown', case=False, na=False).sum())
```

```python
df_clean[df_clean['MOVIES'].str.contains('unknown', case=False, na=False)]
```

```python
df_clean['MOVIES'] = df_clean['MOVIES'].str.strip().replace(r'(?i)^unknown$', np.nan, regex=True)
```

```python
df_clean[df_clean['MOVIES'].str.contains('unknown', case=False, na=False)]
```

```python
df_clean[df_clean['MOVIES'].isna()]
```

```python
#Remove new line characters from the data
string_columns = df_clean.select_dtypes(include = ['object']).columns
print(string_columns)
for col in string_columns:
    df_clean[col] = df_clean[col].str.replace(r'\n+',' ', regex=True)
df_clean.head()
```

```python
#Extract Start Year and End Year from YEAR column
df_clean['START_YEAR'] = df_clean['YEAR'].str.extract(r'(\d{4})').astype(float)
df_clean['END_YEAR'] = df_clean['YEAR'].str.extract(r'–\s?(\d{4})').astype(float)
df_clean.drop(columns = ['YEAR'], inplace=True)
df_clean.head()
```

```python
df_clean['VOTES'] = pd.to_numeric(df_clean['VOTES'].str.replace(',', '', regex=False), errors='coerce')
```

```python
df_clean.head()
```

```python
df_clean['Gross'] = (df_clean['Gross']
                     .str.replace(r'\n+', '', regex=True)
                     .str.strip()
                     .str.replace(r'[\$,]', '', regex=True)
                     .str.replace('M', '')
                     .astype(float) * 1000000)  # if all values are in millions
```

```python
df_clean.head()
```

```python
df_clean[~df_clean['Gross'].isna()]
```

```python
#=====================================================
#STEP 3: DATA TYPE Validation and Conversion
#=====================================================
```

```python
print("\n\n3 Data Type Validation and Conversion")
print("="*50)
```

```python
df_clean['START_YEAR'] = pd.to_numeric(df_clean['START_YEAR'], errors='coerce')
df_clean['END_YEAR'] = pd.to_numeric(df_clean['END_YEAR'], errors='coerce')
df_clean['RunTime'] = pd.to_numeric(df_clean['RunTime'], errors='coerce')
df_clean['VOTES'] = pd.to_numeric(df_clean['VOTES'], errors='coerce')
df_clean['Gross'] = pd.to_numeric(df_clean['Gross'], errors='coerce')
```

```python
df_clean.dtypes
```

```python
print("\n\n4. Derived Columns:")
print("="*50)
```

```python
#Decade
df_clean["decade"] = (df_clean['START_YEAR']//10)*10
#Revenue per minute
df_clean["revenue_per_minute"] = df_clean['Gross'] / df_clean['RunTime']

df_clean.head()
```

```python
print(df_clean.shape)
df_clean = df_clean.dropna(thresh=6)
print(df_clean.shape)
```

```python
df_clean['director'] = df_clean['STARS'].str.extract(r'Director:\s*(.*?)\s*\|')
df_clean['stars'] = df_clean['STARS'].str.extract(r'Stars:\s*(.*)')
df_clean.head()
```
